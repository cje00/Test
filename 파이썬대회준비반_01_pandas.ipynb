### 즐겨먹는 야식 (단위 : %)
'치킨':77,'라면':37,'과자빵':33,'족발보쌈':32,'떡볶이순대':25,'피자':24

```python
# 즐겨먹는 야식 데이터로 시리즈 만들기
# 1차원 리스트로 만들기
s = pd.Series([77,37,33,32,25,24])
s
```

```python
# 인덱스 이름 변경하기
s.index = ['치킨','라면','과자빵','족발보쌈','떡볶이순대','피자']
s
```

```python
# 딕셔너리로 시리즈 생성하기
s = pd.Series({'치킨':77,'라면':37,'과자빵':33,'족발보쌈':32,'떡볶이순대':25,'피자':24})
s
```

```python
# 자료 검색
s['라면']
```

```python
s[['라면','피자']]
```

```python
s[:3]
```

```python
# Series에 요소 추가
s['햄버거'] = 5
s
```

```python
# Series에 요소 삭제
del s['햄버거']
s
```

### csv 파일 불러와서 데이터프레임 만들기
* 데이터위치 (국가통계포털 - 인구총조사 - 인구밀도)
https://kosis.kr/statHtml/statHtml.do?orgId=101&tblId=DT_1B08024&vw_cd=MT_ZTITLE&list_id=A1_13&scrId=&seqNo=&lang_mode=ko&obj_var_id=&itm_id=&conn_path=MT_ZTITLE&path=%252FstatisticsList%252FstatisticsListIndex.do


```python
import pandas as pd
df = pd.read_csv('인구밀도_인구주택총조사기준__20220831220343.csv', encoding='cp949')
df
```

```python
df.info()
```

```python
# 전국 데이터 삭제
df.drop(0,inplace=True)
```

```python
df
```

```python
# 컬럼명 변경하기
df = df.rename(columns = {'행정구역별' : '행정구역'})
```

```python
df
```

```python
# 행정구역별 평균을 구하여 평균 컬럼을 추가
df['평균'] = df.mean(axis=1, numeric_only=True)
#df['평균'] = df.mean(axis=1)
df
```

```python
df.loc['18'] = ['평균', df['2019'].mean(), df['2020'].mean(),df['2021'].mean(), df['평균'].mean()]
```

```python
df.drop(index='18', inplace=True)
df
```

```python
# 인구밀도가 평균이 1000이 넘는 행정구역 
df[df.평균 > 1000].행정구역
```

```python
# 평균이 최소값을 가지는 행정구역
df[df.평균.min() == df.평균].행정구역
```

```python
# 평균 컬럼값으로 정렬
df_sorted = df.sort_values(by='평균', ascending=False)
df_sorted
```

```python
# 인덱스 재설정
df_sorted.reset_index()
#df_sorted.reset_index(drop=True, inplace=True)
```

```python
df_sorted.index =df_sorted.index + 1
df_sorted
```

```python
# 인구밀도가 늘어나고 있는 도시
df[df['2021'] > df.평균 ].행정구역
```
